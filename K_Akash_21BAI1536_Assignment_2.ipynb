```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

Task -2 Load the dataset.

```python
data=pd.read_csv('House Price India.csv')
data.head()
```

```python
data.tail()
```

Task -3 Perform the Below Visualizations.

Univariate Analysis

Histogram

```python
for i in data.columns:
  plt.hist(data[i])
  plt.title(i)
  plt.show()
```

Boxplot

```python
for i in data.columns:
  plt.boxplot(data[i])
  plt.title(i)
  plt.show()
```

Bi Variate Analysis


Scatter Plot

```python
for i in data.columns:
  plt.scatter(data[i],data['Price'])
  plt.xlabel(i)
  plt.ylabel("Price")
  plt.show()
```

Line Chart

```python
for i in data.columns:
  plt.plot(data[i],data['Price'])
  plt.xlabel(i)
  plt.ylabel("Price")
  plt.show()
```

Joinplot

```python
for i in data.columns:
  sns.jointplot(x= data[i],y =data['Price'],data=data)
```

Multivariate Variate Analysis

Pairplot

```python
sns.pairplot(data)
```

Heatmap



```python
sns.heatmap(data.corr())
```
